```python
import json
```

```python
with open('/content/996c937e-ac57-daf068ba3129.geojson') as f:
    problems = json.load(f)
```

```python
big_problems = {}

for i in problems['features']:
    if i['properties']['Причина обращения'] not in big_problems:
        big_problems[i['properties']['Причина обращения']] = 0
    big_problems[i['properties']['Причина обращения']] += 1
```

```python
problems['features'][0]['geometry']['coordinates']
```

```python
from geopy.geocoders import Nominatim
```

```python
geo_loc = Nominatim(user_agent='GetLoc')
loc_name = geo_loc.reverse([59.8506, 30.3209])
```

```python
problem_locations = {}

for i in problems['features']:
    coords = i['geometry']['coordinates']
    coords_key = f'{coords[1]}_{coords[0]}'
    if coords_key not in problem_locations:
        problem_locations[coords_key] = 0
    problem_locations[coords_key] += 1
```

```python
sorted_dict = {k: v for k, v in sorted(problem_locations.items(), key=lambda item: item[1])}
```

```python
for i in list(sorted_dict.keys())[-10:][::-1]:
    coords = i.split('_')
    print(f'{problem_locations[i]} обращений по адресу: {geo_loc.reverse(coords)}')
```

```python
sorted_problems = {k: v for k, v in sorted(big_problems.items(), key=lambda item: item[1])}
```

```python
for i in list(sorted_problems.keys())[-100:][::-1]:
    print(f'{big_problems[i]} обращений типа: {i}')
```

```python
locs_types = []

for i in problem_locations.keys():
    if problem_locations[i] >= 20:
        locs_types.append(i)
```

```python
problem_districts = {}

for i in locs_types:
    coords = i.split('_')
    loc_address = geo_loc.reverse(coords)[0]
    dist_index = loc_address.split(', ').index('Санкт-Петербург') - 1
    if loc_address.split(', ')[dist_index] not in problem_districts:
        problem_districts[loc_address.split(', ')[dist_index]] = []
    problem_districts[loc_address.split(', ')[dist_index]].append(loc_address)
```

```python
n = 1
for i in problem_districts:
    print(f'{n}. Проблемные локации в районе {i}:')
    n_j = 1
    for j in problem_districts[i]:
        print(f'    {n_j}. {j}')
        n_j += 1
    n += 1
```
